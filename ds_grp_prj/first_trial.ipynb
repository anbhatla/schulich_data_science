```python
import pandas as pd

# Load the dataset from the provided link
data_url = "https://raw.githubusercontent.com/anbhatla/schulich_data_science/main/ds_grp_prj/default_of_credit_card_clients.csv"
data = pd.read_csv(data_url)
data.head()
```

```python
# Renaming columns
data.columns = ['ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
                'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
                'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'default_payment_next_month']
data.head()
```

```python
# Checking for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Checking data types
data.dtypes
```

```python
# Summary statistics for numerical columns
numerical_summary = data.describe()
print("Summary statistics for numerical columns:")
print(numerical_summary)

# Summary statistics for categorical columns
categorical_summary = data.describe(include=['object'])
print("\nSummary statistics for categorical columns:")
print(categorical_summary)
```
